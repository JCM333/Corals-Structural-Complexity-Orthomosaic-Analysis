```python
import os
import numpy as np
from PIL import Image
import pylab as plt
```

## Location 1 - Punta Galeras

```python
#Transect 1.A
im_open = Image.open('CT00-Rugosidad.tif', mode ='r') #Load ruggedness index raster
im_array = np.array(im_open)
im_array[im_array < 0] = 0 #remove negative values
CT00 = im_array[1:1501,1:13001] #establishing pixel sample size
CT00_mean = CT00.mean() #mean
CT00_std = CT00.std() #standard deviation
print(CT00_mean, CT00_std)
```

```python
#Visualize Ruggedness Index Transect
plt.matshow(CT00, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 1.B
im_open = Image.open('CT01-Rugosidad.tif', mode ='r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0 #removing negative values
CT01 = im_array[1:1501,1:13001] #establishing pixel sample size
CT01_mean = CT01.mean()
CT01_std = CT01.std()
print(CT01_mean, CT01_std)
```

```python
plt.matshow(CT01, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 1.C
#Raster orientation caused the need to concatenate separate sections for same sample size
im_open = Image.open('CT02-1-Rugosidad.tif', mode = 'r') #First section
im_array1 = np.array(im_open)
im_array1[im_array1 < 0] = 0
print(im_array1.shape)
im_open = Image.open('CT02-2-Rugosidad.tif', mode = 'r') #Second section
im_array2 = np.array(im_open)
im_array2[im_array2 < 0] = 0
print(im_array2.shape)
im_open = Image.open('CT02-3-Rugosidad.tif', mode = 'r') #Third section
im_array3 = np.array(im_open)
im_array3[im_array3 < 0] = 0
print(im_array3.shape)
```

```python
#Establish width and height of array for concatenation
array1_clip = im_array1[1:1502,1:7382]
array2_clip = im_array2[1:1502,1:3576]
array3_clip = im_array3[1:1502,1:2821]

#concatenate
big_array = np.concatenate((array1_clip, array2_clip, array3_clip), axis = 1) 
CT02 = big_array[1:1501,1:13001] #same pixel sample size of built array
#print(CT02.shape)

CT02_mean = CT02.mean()
CT02_std = CT02.std()
print(CT02_mean, CT02_std)
```

```python
plt.matshow(CT02, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 1.D
im_open = Image.open('CT03-Rugosidad.tif', mode = 'r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0
CT03 = im_array[1:1501,1:13001]
CT03_mean = CT03.mean()
CT03_std = CT03.std()
print(CT03_mean, CT03_std)
```

```python
plt.matshow(CT03, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Concatenate all array for Location statistics
CTall = np.concatenate((CT00, CT01, CT02, CT03), axis = 1)
CTall[CTall < 0] = 0

CTall_mean = CTall.mean()
CTall_std = CTall.std()
```

## Location 2 - Punta Pichilingue

```python
#Transect 2.A
im_open = Image.open('PG00-Rugosidad.tif', mode = 'r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0
PG00 = im_array[1:1501,1:10801]
PG00_mean = PG00.mean()
PG00_std = PG00.std()
print(PG00_mean, PG00_std)
```

```python
plt.matshow(PG00, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 2.B
im_open = Image.open('PG01-Rugosidad.tif', mode = 'r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0
im_array = im_array.transpose() #Raster orientation caused the need to transpose array
PG01 = im_array[1:1501,1:10801]
PG01_mean = PG01.mean()
PG01_std = PG01.std()
print(PG01_mean, PG01_std)
```

```python
plt.matshow(PG01, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 2.C
im_open = Image.open('PG02-Rugosidad.tif', mode = 'r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0
PG02 = im_array[1:1501,1:10720]
PG02_mean = PG02.mean()
PG02_std = PG02.std()
print(PG02_mean, PG02_std)
```

```python
plt.matshow(PG02, vmin = 1, vmax=10)
plt.colorbar()
```

```python
#Transect 2.D
im_open = Image.open('PG03-Rugosidad.tif', mode = 'r')
im_array = np.array(im_open)
im_array[im_array < 0] = 0
im_array = im_array.transpose() #Raster orientation caused the need to transpose array
PG03 = im_array[1:1501,1:10801]
PG03_mean = PG03.mean()
PG03_std = PG03.std()
print(PG03_mean, PG03_std)
```

```python
plt.matshow(PG03, vmin = 1, vmax=10)
plt.colorbar()
```

```python
PGall = np.concatenate((PG00, PG01, PG02, PG03), axis = 1)
PGall[PGall < 0] = 0

PGall_mean = PGall.mean()
PGall_std = PGall.std()

print(PGall_mean, PGall_std)
```

## Plots

```python
#Location 1 transect analysis
Transects1 = ['A', 'B', 'C', 'D','Total'] #Names for x
x_pos = np.arange(len(Transects1))

mediasCT = [CT00_mean, CT01_mean, CT02_mean, CT03_mean, CTall_mean] #List of all mean values for Location 1
errorCT = [CT00_std, CT01_std, CT02_std, CT03_std, CTall_std] #List of all standard deviation for Location 1

fig, ax = plt.subplots()
ax.errorbar(x_pos, mediasCT, yerr=errorCT, fmt='o', color="g", markersize=9, alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Ruggedness Index')
ax.set_xticks(x_pos)
plt.ylim(-1, 11)
ax.set_xticklabels(Transects1)
plt.xlabel("Transect")
#ax.set_title('Punta Galeras, B.C.S., MX')
ax.yaxis.grid(True)

#plt.savefig('1_bar_plot_with_error_bars.png')
plt.show()
```

```python
#Location 2 transect analysis
Transects2 = ['A', 'B', 'C', 'D', 'Total']
x_pos = np.arange(len(Transects2))

mediasPG = [PG00_mean, PG01_mean, PG02_mean, PG03_mean, PGall_mean] #List of all mean values for Location 2
errorPG = [PG00_std, PG01_std, PG02_std, PG03_std, PGall_std] #List of all standard deviation for Location 2

fig, ax = plt.subplots()
ax.errorbar(x_pos, mediasPG, yerr=errorPG, fmt='o', color="g", markersize=9, alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Ruggedness Index')
ax.set_xticks(x_pos)
plt.ylim(-0.5, 11)
ax.set_xticklabels(Transects2)
plt.xlabel("Transects")
#ax.set_title('Punta Pichilingue, B.C.S.')
ax.yaxis.grid(True)

#plt.savefig('2_bar_plot_with_error_bars.png')
plt.show()
```

## One-Way ANOVA

```python
#Statistics library
import scipy.stats

#Sample size and vector conversion required for one-way ANOVA
Location1 = CTall[0:1500,0:43000].ravel()
Location2 = PGall[0:1500,0:43000].ravel()

#Analysis
oneway = scipy.stats.f_oneway(Location1,Location2)
print(oneway)
```

```python
#Box Plot
data = [Location1, Location2]

f, (ax2, ax) = plt.subplots(2, 1, sharex=True)

ax.boxplot(data)
ax2.boxplot(data)

ax2.set_ylim(60, 140)
ax.set_ylim(0, 15)

ax.spines['top'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax.xaxis.tick_top()
ax2.tick_params(labeltop=False)
ax.xaxis.tick_bottom()

#plt.savefig('BoxOneWay.png', dpi=100)
plt.show()
```
